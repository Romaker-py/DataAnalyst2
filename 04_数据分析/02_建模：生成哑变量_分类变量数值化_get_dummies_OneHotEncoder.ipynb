#### 不同函数说明

sklearn.preprocessing.OrdinalEncoder()：把X值中的分类变量编码为数值型有序变量。

sklearn.preprocessing.OneHotEncoder()：把X值中的分类变量编码为哑变量。【不用于y值】

sklearn.preprocessing.LabelEncoder()：把y值（可以是数值型和字符型）编码为数值型有序变量。

sklearn.preprocessing.LabelBinarizer()：把y值中多分类变量编码为二分类变量。【用于y值】

#### 从数据库导入数据

```python
import pymysql.cursors
import pandas as pd

db = pymysql.connect(host='localhost',user='root',password='root',db='lagou',charset='utf8',cursorclass=pymysql.cursors.DictCursor)
cursor = db.cursor()
sql = 'SELECT * FROM dataset;'
cursor.execute(sql)
result = cursor.fetchall()
df = pd.DataFrame(list(result))
db.close()

df
```

### 一、使用pd.get_dummies()

【优点】只转换分类变量，原数值变量继续保留——符合预期。

```python
import pandas as pd

df_dummies = pd.get_dummies(df)
df_dummies
```

### 二、使用sklearn.preprocessing.OneHotEncoder()

【不足】原来的数值变量也一起转换；新的变量名看不出字段含义。

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

enc = OneHotEncoder(dtype=np.int, sparse=False)
```

```python
array = enc.fit_transform(df)
columns = enc.get_feature_names()
df_enc = pd.DataFrame(array, columns=columns)
df_enc
```
