#### 从数据库导入数据

```python
import pymysql.cursors
import pandas as pd

db = pymysql.connect(host='localhost',user='root',password='root',db='lagou',charset='utf8',cursorclass=pymysql.cursors.DictCursor)
cursor = db.cursor()
sql = 'SELECT * FROM dataset;'
cursor.execute(sql)
result = cursor.fetchall()
df = pd.DataFrame(list(result))
db.close()

df
```

#### 拆分训练集和测试集

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-1]
X = pd.get_dummies(X).values
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_names = pd.get_dummies(df.iloc[:,:-1]).columns
X_names
```

#### 堆积分类器建模

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import VotingClassifier

dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
print('DecisionTreeClassifier得分为：%0.3f' % dtc.score(X_test, y_test))
print()

rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
print('RandomForestClassifier得分为：%0.3f' % rfc.score(X_test, y_test))
print()

gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train, y_train)
print('GradientBoostingClassifier得分为：%0.3f' %gbc.score(X_test, y_test))
print()

xgbc = XGBClassifier(random_state=42)
xgbc.fit(X_train, y_train)
print('XGBClassifier得分为：%0.3f' % xgbc.score(X_test, y_test))
print()

lgbmc = LGBMClassifier(random_state=42)
lgbmc.fit(X_train, y_train)
print('LGBMClassifier得分为：%0.3f' % lgbmc.score(X_test, y_test))
print()

abc = AdaBoostClassifier(random_state=42)
abc.fit(X_train, y_train)
print('AdaBoostClassifier得分为：%0.3f' % abc.score(X_test, y_test))
print()

lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
print('LogisticRegression得分为：%0.3f' % lr.score(X_test, y_test))
print()

lrcv = LogisticRegressionCV(random_state=42)
lrcv.fit(X_train, y_train)
print('LogisticRegressionCV得分为：%0.3f' % lrcv.score(X_test, y_test))
print()

mlpc = MLPClassifier(random_state=42)
mlpc.fit(X_train, y_train)
print('MLPClassifier得分为：%0.3f' %mlpc.score(X_test, y_test))
print()

svc = SVC(random_state=42)
svc.fit(X_train, y_train)
print('SVC得分为：%0.3f' % svc.score(X_test, y_test))
print()

lsvc = LinearSVC(random_state=42)
lsvc.fit(X_train, y_train)
print('LinearSVC得分为：%0.3f' % lsvc.score(X_test, y_test))
print()

knc = KNeighborsClassifier()    # 无random_state参数
knc.fit(X_train, y_train)
print('KNeighborsClassifier得分为：%0.3f' % knc.score(X_test, y_test))
print()

gnb = GaussianNB()    # 无random_state参数
gnb.fit(X_train, y_train)
print('GaussianNB得分为：%0.3f' % gnb.score(X_test, y_test))
print()

estimators = [
    ('dtc',DecisionTreeClassifier(random_state=42)),
    ('rfc',RandomForestClassifier(random_state=42)),
    ('gbc',GradientBoostingClassifier(random_state=42)),
    ('xgbc',XGBClassifier(random_state=42)),
    ('lgbmc',LGBMClassifier(random_state=42)),
    ('abc',AdaBoostClassifier(random_state=42)),
    ('lr',LogisticRegression(random_state=42)),
    ('lrcv',LogisticRegressionCV(random_state=42)),
    ('mlpc',MLPClassifier(random_state=42)),
    ('svc',SVC(random_state=42)),
    ('lsvc',LinearSVC(random_state=42)),
    ('knc',KNeighborsClassifier()),
    ('gnb',GaussianNB())
]
clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(random_state=42))
clf.fit(X_train, y_train)
print('StackingClassifier得分为：%0.3f' % clf.score(X_test, y_test))
print()

vc = VotingClassifier(estimators=[
    ('dtc',dtc),
    ('rfc',rfc),
    ('gbc',gbc),
    ('xgbc',xgbc),
    ('lgbmc',lgbmc),
    ('abc',abc),
    ('lr',lr),
    ('lrcv',lrcv),
    ('mlpc',mlpc),
    ('svc',svc),
    ('lsvc',lsvc),
    ('knc',knc),
    ('gnb',gnb)
], n_jobs=-1)
vc.fit(X_train, y_train)
print('VotingClassifier得分为：%0.3f' % vc.score(X_test, y_test))
print()
```
