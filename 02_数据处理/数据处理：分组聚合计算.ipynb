```python
import pandas as pd

df = pd.read_csv(r'D:\Jupyter\Data\smsReplyRate_2days.csv',encoding='gb18030')
df
```

### 一、group by

#### 一对多

```python
df.groupby('staffId').sum()
```

#### 一对一

```python
df.groupby('staffId')['smsTotal'].sum()
```

#### 多对一

```python
df.groupby(['staffId','company'])['smsTotal'].sum()
```

#### 多对多

```python
df.groupby(['staffId','company','subCompany']).sum()
```

#### 多种聚合

```python
df.groupby(['staffId','company','subCompany']).agg({'smsTotal':'sum',
                                                    'replySpeed':'mean',
                                                    'threeMinNum':'sum',
                                                    'threeToSixNum':'sum',
                                                    'notReplyNum':'sum'})
```
