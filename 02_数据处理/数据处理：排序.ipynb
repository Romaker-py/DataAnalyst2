```python
import pandas as pd

df = pd.read_csv(r'D:\Jupyter\Data\auto_ins.csv',encoding='gb2312')
df
```

### 一、对DataFrame排序

```python
?df.sort_values
```

#### 根据指定字段排序

```python
df.sort_values(by='Age')    # 想要排序生效，inplace改为True即可
```

#### 根据指定字段降序排序

```python
df.sort_values(by='Age',ascending=False)    # 想要排序生效，inplace改为True即可
```

#### 根据多个字段排序

```python
df.sort_values(by=['Age','exp'])    # 想要排序生效，inplace改为True即可
```

#### 对多个字段升序和降序排序

```python
df.sort_values(by=['Age','exp'],ascending=(True,False))    # 想要排序生效，inplace改为True即可
```

#### 把错误值排到最前面

```python
df.sort_values(by='Age',na_position='first')    # 想要排序生效，inplace改为True即可
```

### 二、对DataFrame的字段值排序

##### 对值排序

```python
df['Age'].value_counts().sort_values()
```
