### 二、删除字段

方法：
1、直接用del删除指定字段；

2、使用drop删除：
①使用df重新赋值的方式删除；
②用columns指定删除的列；
③用axis=1删除列。

#### 使用del删除字段

// 一次删除一个字段：

```python
del df['employee']
df
```

// 一次删除多个字段：

```python
df['aaa'], df['bbb'] = True, False
df
```

```python
del df['aaa'], df['bbb']
df
```

#### 使用drop删除字段（方式一）

// 一次删除一个字段：

```python
df['employee'] = True
df
```

```python
df.drop('employee', axis=1, inplace=True)    # 注意：inplace默认为False，要手动修改【axis参数为必选】
df
```

// 一次删除多个字段：

```python
df['aaa'], df['bbb'] = True, False
df
```

```python
df.drop(['aaa', 'bbb'], axis=1, inplace=True)
df
```

#### 使用drop删除字段（方式二）

```python
df['employee'] = True
df
```

```python
df.drop(columns=['employee'],inplace=True)
df
```

#### 使用drop删除字段（方式三）

```python
df['employee'] = True
df
```

```python
df = df.drop(columns=['employee'])
df
```

### 四、修改字段名称

```python
import pandas as pd
df = pd.DataFrame([[101,'frank', 'M', 29], [102,'mary', 'F', 23], [103,'tom', 'M', 35], [104,'ted', 'M', 33], [105,'jean', 'F', 21], [106,'lisa', 'F', 20]],
                  columns = ['userid','name', 'gender', 'age'])
df
```

#### 方式一：rename

// 修改单个字段：

```python
df.rename(columns={'gender':'sex'},inplace=True)
df
```

// 修改多个字段：

```python
df.rename(columns={'gender':'sex', 'age':'birthyear'},inplace=True)
df
```

#### 方式二：list

另外，list里面也可以是另外一个DataFrame的列，即df2.columns。

```python
df.columns = list(['userid', 'name', 'gender', 'age'])
df
```

### 五、调整字段顺序

```python
order = ['userid', 'name', 'age', 'gender']
df = df[order]
df
```
