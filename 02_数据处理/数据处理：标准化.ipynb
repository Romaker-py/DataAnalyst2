#### Z-Score标准化

这里直接用map函数来计算标准化。如果调用sklearn的preprocessing.StandardScale来标准化，需要进行额外的处理，否则会报错。

```python
df['Age_zscore'] = df['Age'].map(lambda x: (x-df['Age'].mean())/df['Age'].std())
df
```
