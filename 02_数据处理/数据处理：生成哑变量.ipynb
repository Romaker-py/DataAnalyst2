```python
import pandas as pd

df = pd.read_csv(r'D:\Jupyter\Data\auto_ins.csv',encoding='gb2312')
df
```

#### 单个生成哑变量

```python
?pd.get_dummies
```

```python
Owner_merge = pd.get_dummies(df['Owner'],prefix='Owner',drop_first=False)    # drop_first如果把默认的值改为True，则会去掉第一个哑变量
Owner_merge
```

##### 合并原df和新的哑变量

pd.concat合并后只保留一个index

```python
df = pd.concat([df,Owner_merge],axis=1)
df
```

#### 批量生成哑变量（把字符变量转成数值变量）

```python
df2 = pd.get_dummies(df)
df2
```
