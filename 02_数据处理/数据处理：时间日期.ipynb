### 一、转换成日期格式

#### 方式一：to_datetime

```python
df['birthday'] = pd.to_datetime(df['birthday'])
```

```python
df.dtypes
```

#### 方式二：lambda

```python
from datetime import datetime
df['birthday'] = df['birthday'].map(lambda x: datetime.strptime(str(x),'%Y%m%d'))
```

```python
df
```

```python
df.info()
```

### 二、日期间隔计算

获取当前日期

```python
from datetime import date
date.today()
```

获取当前时间

```python
datetime.now()
```

```python
df['dif'] = df['birthday'].map(lambda x: (datetime.now()-x).days)
df
```

### 三、日期加减

日期加7天

```python
import datetime
df['oneWeekAfter'] = df['birthday'].map(lambda x: x+datetime.timedelta(days=7))
df
```

日期减7天

```python
import datetime
df['oneWeekBefore'] = df['birthday'].map(lambda x: x-datetime.timedelta(days=7))
df
```

### 四、输出日期字符串

```python
import time
from datetime import datetime

print(datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
print(time.strftime('%Y-%m-%d %H:%M:%S'))
print(time.strftime("%Y{}%m{}%d{} %H{}%M{}%S{}",time.localtime()).format('年','月','日','时','分','秒'))
print(datetime.now().strftime("%Y{}%m{}%d{} %H{}%M{}%S{}").format('年','月','日','时','分','秒'))
```

### 七、日期统计数据

#### 按周统计

```python
df_stat = df.resample('w').sum()
df_stat
```

#### 按月统计

```python
df_stat = df.resample('M').sum()    # "MS"是每个月第一天为开始日期, "M"是每个月最后一天
df_stat
```

#### 按季度统计

```python
df_stat = df.resample('Q').sum()    # "QS"是每个季度第一天为开始日期, "Q"是每个季度最后一天
df_stat
```

#### 按年度统计

```python
df_stat = df.resample('A').sum()    # "AS"是每年第一天为开始日期, "A是每年最后一天
df_stat
```

### 八、按日期统计显示

#### 按季度统计并显示

```python
df_stat = df.resample('Q').sum().to_period('Q')
df_stat
```

#### 按月度统计并显示

### 九、提取指定类型日期

#### 从datetime里面提取

```python
import time

time.localtime()
```

```python
from datetime import datetime

datetime.now()
```

```python
import pandas as pd

df_time = pd.DataFrame([datetime.now()], columns=['datetime'])
df_time
```

```python
datetime = df_time['datetime'].iloc[0]

print('年份为：%s' % datetime.year)
print('月份为：%s' % datetime.month)
print('日为：%s' % datetime.day)
print('星期为【weekday()+1】：%s' % (datetime.weekday() + 1))    # 注意：星期需要加括号。原周一到周日为0-6，现改为1-7
print('星期为【dayofweek+1】：%s' % (datetime.dayofweek + 1))    # 注意：原周一到周日为0-6，现在为1-7
print('时为：%s' % datetime.hour)
print('分为：%s' % datetime.minute)
print('秒为：%s' % datetime.second)
print('微秒（百万分之一秒）为：%s' % datetime.microsecond)
print('年度第几周为：%s' % datetime.week)
```
