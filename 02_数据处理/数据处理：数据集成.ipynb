##### 导入数据

```python
import pandas as pd

df = pd.read_csv(r'D:\Jupyter\Data\auto_ins.csv',encoding='gb2312')
df
```

##### 生成哑变量

```python
Owner_dummies = pd.get_dummies(df['Owner'],prefix='Owner',drop_first=False)    # drop_first如果把默认的值改为True，则会去掉第一个哑变量
Owner_dummies
```

##### 创建一个DataFrame

```python
df_sex = pd.DataFrame([['男',1],['女',0]],columns=['sex','is_men'])
df_sex
```

## 总结

#### 1、合并字段

1、两张表index顺序和长度完全一致时，用pd.concat【index自动去重】；

2、两张表长度和index均不一样时，用join，且右表需要set_index，把共同字段（字段名称可以不一样）设置为索引【index自动去重】。

#### 2、合并行

可以使用append【比较推荐：符合习惯，不易出错】，或者concat。

====================================================================================================================

### 一、合并列

#### 合并两个不同字段，但相同index的DataFrame【推荐使用pd.concat】

注意：使用pd.concat时，需要把待合并的两个DataFrame放到数组[ ]里面。

```python
?pd.concat
```

```python
df_Owner_concat = pd.concat([df,Owner_dummies],axis=1)
df_Owner_concat
```

#### 合并两个不同字段，但相同index的DataFrame【使用merge】——待去重

```python
df_Owner_merge = df.merge(Owner_dummies,on=df.index)
df_Owner_merge
```

#### 合并两个不同字段的DataFrame【使用join】

```python
?df.join
```

【说明】
1、set_index才能保证左边df的index不受影响；
2、on的值为左边df的字段；
3、两个表的键名称可以不相同。

```python
df_sex_join = df.join(df_sex.set_index('sex'),on='Gender')
df_sex_join
```

#### 合并两个不同字段的DataFrame【使用merge】——有待去重

##### 方式一（官方示例）

```python
?df.merge
```

```python
df_sex_merge = df.merge(df_sex,left_on='Gender',right_on='sex')
df_sex_merge
```

##### 方式二——有待去重

```python
df_sex_merge = pd.merge(df,df_sex,left_on='Gender',right_on='sex')
df_sex_merge
```
