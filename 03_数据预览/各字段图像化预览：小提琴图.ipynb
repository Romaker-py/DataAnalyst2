```python
import pandas as pd

df = pd.read_csv(r'D:\Jupyter\Data\auto_ins.csv',encoding='gb2312')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置-黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
sns.set(font='SimHei')  # 解决Seaborn中文显示问题
#sns.set(style="ticks")    # 风格：white, dark, whitegrid(默认), darkgrid, ticks

plt.figure(figsize=(12,8))
sns.violinplot(data=df,x='Gender',y='Age',palette='Set2',hue='Garage',split=True,order=['女','男'])  # hue为二分类变量，split决定是否合并显示
plt.title('小提琴图')
#plt.show()    # show之后，无法导出图片
plt.savefig(r"D:\Jupyter\Data\violinplot.png",dpi=600)
```
