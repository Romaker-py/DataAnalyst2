### 一、Map

将函数套用到Series上的每个元素（elementwise）

官方说明：Series.map: For element-wise operations.

#### 举例：对Loss进行标准化

```python
df['Loss_zscore'] = df['Loss'].map(lambda x: (x-df['Loss'].mean())/df['Loss'].std())
df
```

### 二、Apply

_注意事项：axis=1不可少_

将函数套用到DataFrame上的行和列

官方说明：DataFrame.apply : Apply a function row-/column-wise.

#### 举例：求每位学生三次考试成绩中，最高分和最低分的差值

```python
score = pd.DataFrame([
                    [60,70,50],\
                    [80,79,68],\
                    [63,66,82]], columns = ['First', 'Second', 'Third'])
score
```

```python
score['dif'] = score.apply(lambda x: x.max()-x.min(),axis=1)
score
```

#### 也可以在指定的几列中进行求差值

```python
score['dif'] = score[['First','Second','Third']].apply(lambda x: x.max()-x.min(),axis=1)
score
```
