```python
import pandas as pd
df = pd.read_csv(r'D:\Jupyter\Data\【拉勾网】数据分析-深圳（2020年03月13日 14时49分22秒）.csv',
                 encoding='gb18030',parse_dates=['createTime','lastLogin','setupDate'])
df
```

```python
from sqlalchemy.types import *
from sqlalchemy import create_engine
# pandas类型和sql类型转换
def map_types(df):
    dtypedict = {}
    for i, j in zip(df.columns, df.dtypes):
        # 岗位描述和公司简介特别长
        #if i == 'jobDetail' or i == 'companyProfile':
        if i in ['jobDetail','companyProfile']:
            dtypedict.update({i: Text()})
        elif i in ['latitude','longitude','matchScore']:
            dtypedict.update({i: DECIMAL(9,6)})    # DECIMAL要全部大写【左参数：整数和小数的总位数，右参数：小数位数】
        elif i in ['positionId','companyId','publisherId']:
            dtypedict.update({i: NVARCHAR(length=20)})
        elif i in ['setupDate']:
            dtypedict.update({i:Date()})
        elif "object" in str(j):
            dtypedict.update({i: NVARCHAR(length=255)})
        elif "float" in str(j):
            dtypedict.update({i: Float()})    # precision=2, asdecimal=True
        elif "int" in str(j):
            dtypedict.update({i: Integer()})
        elif 'bool' in str(j):
            dtypedict.update({i: Boolean()})
    return dtypedict

# 通过dtype设置类型 为dict格式{“col_name”:type}
dtypedict = map_types(df)


# 值得注意的是，需要提前创建lagou数据库
engine = create_engine('mysql+pymysql://root:root@localhost:3306/lagou?charset=UTF8MB4')
con = engine.connect()
# 表名可以不提前创建
df.to_sql(name='data_analyst', con=con, if_exists='replace', index=False, dtype=dtypedict)
# 真正关掉连接需要用dispose()，不能用close()
con.close()
engine.dispose()
print('成功导入MySQL！')
```
