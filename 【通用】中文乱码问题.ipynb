### 三、sklearn的DecisionTreeClassifier算法graphviz保存图片乱码

参考链接：Graphviz图片显示中文乱码问题 https://www.cnblogs.com/Maggieli/p/12113772.html

```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
wine = load_wine()
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size=0.3)
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest) 
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']
```

```python
import graphviz
dot_data = tree.export_graphviz(clf
                               ,out_file = None
                               ,feature_names = feature_name
                               ,class_names=["琴酒","雪莉","贝尔摩德"]
                               ,filled=True   
                               ,rounded=True  
)
graph = graphviz.Source(dot_data)
graph.render('D:/Jupyter/Data/result',format='png',cleanup=True)
```

导出的文件显示中文是乱码的，解决方法如下：

```python
import graphviz
dot_data = tree.export_graphviz(clf
                               ,out_file = 'D:/Jupyter/Data/result.dot'
                               ,feature_names = feature_name
                               ,class_names=["琴酒","雪莉","贝尔摩德"]
                               ,filled=True   
                               ,rounded=True  

)
with open('D:/Jupyter/Data/result.dot',encoding='utf-8') as f:
    dot_graph = f.read()
graph=graphviz.Source(dot_graph.replace("helvetica","SimHei"))
graph.render('D:/Jupyter/Data/result',format='png',cleanup=True)
```
